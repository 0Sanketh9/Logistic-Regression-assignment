```python
import pandas as pd 
import numpy as np 
import seaborn as sp 
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import  metrics
from sklearn.metrics import classification_report
```

```python
bank=pd.read_csv('bank.csv')
bank
```

```python
bank.info()
```

```python
#gives null values present in columns
bank.isnull().sum()
```

```python
bank.describe
```

```python
# One-Hot Encoding of categrical variables
bank1=pd.get_dummies(bank,columns=['job','marital','education','contact','poutcome'])
bank1
```

```python
# To see all columns
pd.set_option("display.max.columns", None)
bank1
```

```python
# Custom Binary Encoding of Binary o/p variables 
bank1['default'] = np.where(bank1['default'].str.contains("yes"), 1, 0)
bank1['housing'] = np.where(bank1['housing'].str.contains("yes"), 1, 0)
bank1['loan'] = np.where(bank1['loan'].str.contains("yes"), 1, 0)
bank1['y'] = np.where(bank1['y'].str.contains("yes"), 1, 0)
bank1
```

```python
bank1['month'].value_counts()
```

```python
order={'month':{'jan':1,'feb':2,'mar':3,'apr':4,'may':5,'jun':6,'jul':7,'aug':8,'sep':9,'oct':10,'nov':11,'dec':12}}
```

```python
bank1=bank1.replace(order)
bank1
```

```python
bank1.info()
```

```python
bank1.isnull().sum()
```

```python
bank1.shape
```

```python
#Model building 
x=pd.concat([bank1.iloc[:,0:11],bank1.iloc[:,12:]],axis=1)
y=bank1.iloc[:,11]
classifier=LogisticRegression()
classifier.fit(x,y)
```

```python
classifier.coef_ 
```

```python
classifier.predict(x)     #gives probaility values of X
```

```python
y_pred=classifier.predict(x)    #gives probability values of x
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'y_pred_prob':y_pred})
y_pred_df
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y,y_pred)
print (confusion_matrix) 
```

```python
pd.crosstab(y_pred,y)
```

```python
# The model accuracy is calculated by (a+d)/(a+b+c+d)
(39107+1282)/(39107+815+4007+1282)
```

```python
accuracy = sum(y==y_pred)/bank1.shape[0]
accuracy 
```

```python
print (classification_report (y, y_pred))     # y=actual value ,y_pred=predicted value
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
